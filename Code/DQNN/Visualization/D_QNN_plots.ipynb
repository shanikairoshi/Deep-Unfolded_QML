```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
iris_df = pd.read_csv('/content/training_resultsIrisDQNN.csv')
adhoc_df = pd.read_csv('/content/training_resultsAdhocDQNN.csv')
genomic_df = pd.read_csv('/content/training_resultsGenomicDQNN.csv')
breastcancer_df = pd.read_csv('/content/training_results_BreatCancerDQNN.csv')

# Dictionary to hold the dataframes
datasets = {
    "Iris": iris_df,
    "Adhoc": adhoc_df,
    "Genomic": genomic_df,
    "BreastCancer": breastcancer_df
}

# Plot 1: Training Accuracy
plt.figure(figsize=(10, 5))
for dataset_name, df in datasets.items():
    plt.plot(df['Iteration'], df['Training Accuracy'], label=dataset_name)
plt.title('Training Accuracy vs Iterations')
plt.xlabel('Iteration')
plt.ylabel('Training Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

# Plot 2: Test Accuracy
plt.figure(figsize=(10, 5))
for dataset_name, df in datasets.items():
    plt.plot(df['Iteration'], df['Test Accuracy'], label=dataset_name)
plt.title('Test Accuracy vs Iterations')
plt.xlabel('Iteration')
plt.ylabel('Test Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

# Plot 3: Objective Function Value
plt.figure(figsize=(10, 5))
for dataset_name, df in datasets.items():
    plt.plot(df['Iteration'], df['Objective Function Value'], label=dataset_name)
plt.title('Objective Function Value vs Iterations')
plt.xlabel('Iteration')
plt.ylabel('Objective Function Value')
plt.legend()
plt.tight_layout()
plt.show()
```
