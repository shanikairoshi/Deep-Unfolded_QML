```python
import csv
import matplotlib.pyplot as plt

def read_optimizer_log(filename):
    evaluations = []
    costs = []

    with open(filename, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            evaluations.append(int(row['Evaluation']))
            costs.append(float(row['Cost']))

    return evaluations, costs

# Read data from CSV files
evals_gd, costs_gd = read_optimizer_log('optimizer_log_GradientDescent.csv')
evals_dun_gd, costs_dun_gd = read_optimizer_log('optimizer_log_dun.csv')
evals_dun_gd_Learn, costs_dun_gd_Learn = read_optimizer_log('optimizer_log_dun_Learn.csv')

# Plot the cost over evaluations for GradientDescent
plt.figure(figsize=(10, 6))
plt.plot(evals_gd, costs_gd, label='GradientDescent', color='blue')
plt.plot(evals_dun_gd, costs_dun_gd, label='Deep Unfolding GradientDescent', color='red')
#plt.plot(evals_dun_gd_Learn, costs_dun_gd_Learn, label='Deep Unfolding GradientDescent Learn', color='green')


plt.xlabel('Evaluation')
plt.ylabel('Cost')
plt.title('Cost vs. Evaluation Iris data for VQC GradientDescent and VQC Deep Unfolding GradientDescent')
plt.legend()
plt.grid(True)
plt.show()
```
